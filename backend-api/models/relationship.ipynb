

This notebook explains how entity relationships are represented in code using **1-to-many**, **many-to-one**, and **composite key (bridge table)** relationships.

## 1. One-to-Many (1-to-M)
One parent can have many children. The parent has a `ICollection<T>` and the child has a foreign key.

```csharp
// Category → Products (1-to-M)
public class Category
{
    public int Id { get; set; }
    public string Name { get; set; }
    public ICollection<Product> Products { get; set; } // One category has many products
}

public class Product
{
    public int Id { get; set; }
    public string Name { get; set; }
    public int CategoryId { get; set; }              // Belongs to one category
    public Category Category { get; set; }
}
```

## 2. Many-to-One (M-to-1)
The child has a foreign key pointing to the parent. Example: Many products → One supplier.

```csharp
public class Supplier
{
    public int Id { get; set; }
    public string Name { get; set; }
    public ICollection<Product> Products { get; set; } // One supplier has many products
}

public class Product
{
    public int SupplierId { get; set; }              // Belongs to one supplier
    public Supplier Supplier { get; set; }
}
```

## 3. Customer → Orders (1-to-M)
One customer can place many orders. Each order belongs to one customer.

```csharp
public class Customer
{
    public int Id { get; set; }
    public string FirstName { get; set; }
    public ICollection<Order> Orders { get; set; } // One customer has many orders
}

public class Order
{
    public int Id { get; set; }
    public int CustomerId { get; set; }           // Belongs to one customer
    public Customer Customer { get; set; }
}
```

## 4. Order → OrderItems (1-to-M)
One order has many order items.

```csharp
public class Order
{
    public int Id { get; set; }
    public ICollection<OrderItem> OrderItems { get; set; } // One order has many order items
}

public class OrderItem
{
    public int OrderId { get; set; }              // Belongs to one order
    public Order Order { get; set; }
}
```

## 5. Product → OrderItems (1-to-M)
One product can appear in many order items.

```csharp
public class Product
{
    public int Id { get; set; }
    public ICollection<OrderItem> OrderItems { get; set; } // Product in many order items
}

public class OrderItem
{
    public int ProductId { get; set; }            // Belongs to one product
    public Product Product { get; set; }
}
```

## 6. OrderItem as a Composite Key (Bridge Table)
OrderItem connects Order and Product to represent a shopping cart line.
- Composite Key: OrderId + ProductId
- Prevents same product repeating in one order.

```csharp
public class OrderItem
{
    public int OrderId { get; set; }
    public int ProductId { get; set; }
    public int Quantity { get; set; }
    public decimal UnitPrice { get; set; }
}
```

## Summary
1. `ICollection<T>` in parent = has many
2. `int ParentId` in child = belongs to one
3. Composite key in linking table = connects two entities many-to-many
4. This pattern justifies all relationships in code clearly.
